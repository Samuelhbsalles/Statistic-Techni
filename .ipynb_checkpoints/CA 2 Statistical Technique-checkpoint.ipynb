# Explation regarding to perform One-Sample t-test:

1. When we have a population mean, which is 70%. I can perform One-sample T-test to be able to see if my population mean is different from the population mean. 
2. The one-sample T-test assumes that the distribution of population is normal, if my sample are normally distributed or not. and as we can see in the plot 1 is normally distributed. 


#### Import Libraries

```python
import scipy.stats as stats
```

```python
# The Parameters. 
```

```python
alpha = 0.05
national_average = 68.73
stats.ttest_1samp(a=df, popmean=0.7)
```

```python
p_value = 1.9465122
```

```python
plt.bar(['Dublin Students', 'National Average'], [df['exam_score'].mean(), national_average], color=['blue', 'red'])
plt.xlabel('Group')
plt.ylabel('Mean Exam Scores')
plt.title('Comparison of Dublin Students\' Mean Exam Scores to National Average')
plt.show()
```

```python
# Create a function to see if is justified the concerns or not. 
if p_value < alpha:
    print("\nThe concerns are justified. The performance of the students is below the national average.")
else:
    print("\nThe concerns are not justified. There is no evidence that the performance is below the national average.")
```

# Dataset Diamonds 

### In this case the best approach would be chi2_contingency due to find the associations between two categorical variables. 
   In this exercice we have the alpha = 0.01 and also were tasked to find and interpret 90% confidence intervals for both the mean price of colourless diamonds and the mean price of non-colourless diamonds.


2. Load the diamonds dataset, and print the first 5 rows. The color variable refers to the colour of the 
diamond, with categories from D to J. Colourless diamonds are considered better than 
diamonds with a yellow tint. Diamonds from D to F are considered colourless, and diamonds 
from colour G to J are not considered colourless (that is, they have a very faint colour).
1. Create a new binary variable in the dataframe called colourless which records 1 in rows 
    with colourless diamonds and 0 otherwise.
2. Perform an appropriate hypothesis test to determine whether there is any association 
    between the clarity of a diamond and whether it is colourless or not. Use a significance 
    level of alpha = 0.01.
3. Produce and comment on an appropriate plot to illustrate your findings. 
4. Find and interpret 90% confidence intervals for both the mean price of colourless diamonds 
    and the mean price of non-colourless diamonds.

```python
import pandas as pd
from scipy.stats import chi2_contingency
import seaborn as sns
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
df1 = pd.read_csv('diamonds.csv')
```

```python
df1.head()
```

```python
df1.color.value_counts(dropna=False)
```

```python
df1['colourless'] = (df1['color'] <= 'F').astype(int)
# Creating a new variable binary call colourless. 
```

```python
df1.head()
```

```python
df1.shape
```

```python
import scipy.stats as stats
```

```python
ct = pd.crosstab(df1['clarity'], df1['colourless'])
```

```python
chi2, p, dof, expected = chi2_contingency(ct)
```

```python
# Create a function to verify if we reject the null hypothesis or not. 
alpha = 0.01
if p <= alpha:
    print("Reject the null hypothesis. There is a significant association.")
else:
    print("Fail to reject the null hypothesis. There is no significant association.")
    
```

```python
sns.countplot(x='clarity', hue='colourless', data=df1)
plt.title('Association between Clarity and Colourless Diamonds')
plt.show()
```

```python
import numpy as np
```

```python
conf_int1 = df1[df1['colourless'] == 1]['price'].mean()
conf_int_non2 = df1[df1['colourless'] == 0]['price'].mean()

print(f"90% Confidence Interval of colourless: {conf_int1}")
print(f"90% Confidence Interval of Non-Colourless : {conf_int_non2}")
```

# Dataset PlantGrowth

1. Find and comment on important summary statistics by treatment and produce an 
    appropriate plot to summarise the dataset.
2. Conduct an appropriate hypothesis test to see if there is evidence of a difference between 
    the three means (that is, the control and the two treatments). Use a significance level of 
    alpha = 0.05. 
3. If there is evidence of a difference between the three means, find and comment on where 
    this difference may be.


```python
pip install pydataset
```

```python
from pydataset import data
from scipy.stats import f_oneway
```

```python
df2 = data('PlantGrowth')
```

```python
df2.head()
```

```python
df2.shape
```

```python
df2.describe()
```

```python
sns.boxplot(x='group', y='weight', data=df2)
plt.title('Plant Growth by Treatment')
plt.xlabel('Treatment Group')
plt.ylabel('Plant Weight')
plt.show()
```

#### The box plot suggests that treatment group 1 trt1 has the highest median in plant weight, however, the least in variability in planta weight.
It suggests that treatment group 1 may be more effective. 

```python
control = df2[df2['group'] == 'ctrl']['weight']
treatment1 = df2[df2['group'] == 'trt1']['weight']
treatment2 = df2[df2['group'] == 'trt2']['weight']
```

```python
f_statistic, p_value = f_oneway(control, treatment1, treatment2)
```

```python
print(f_statistic, p_value)
```

```python
alpha = 0.05 
if p_value < alpha: 
    print('There is evidence related to the difference between the means')
else:
    print('There is no evidence related to the difference between the means')
    
    
```

#### If the p-value is less than alpha, We can conclude that there is evidence of a significant difference between the means. Otherwise, we would fail to reject the null hypothesis. 

```python
# Learn about it 

from statsmodels.stats.multicomp import pairwise_tukeyhsd

# Perform Tukey's HSD post-hoc test
tukey_results = pairwise_tukeyhsd(df2['weight'], df2['group'])

# Display the summary 
print(tukey_results.summary())

# Plot the results
tukey_results.plot_simultaneous(comparison_name='ctrl', xlabel='Plant Weight')
plt.show()

```

```python
import matplotlib.pyplot as plt 

plt.hist(control, alpha=0.5, label='ctrl' )
plt.hist(treatment1, alpha=0.5, label='trt1')
plt.hist(treatment2, alpha=0.5, label='trt2')
#plot title and axis labels
plt.title("Test")
plt.xlabel("Test")
plt.ylabel("Test")

plt.legend(title = "CA1 TEST")

plt.show
```

# Dataset trees

   Load the trees dataset from the pydataset library. It contains measurements of the diameter, 
   height and volume of timber in 31 felled black cherry trees. Note that the diameter (in inches) is 
   labelled girth in the dataset. It is measured at 4 foot 6 inches above the ground.
1. Perform a correlation analysis between all numerical variables. Include and comment on 
    the results of hypothesis tests for the population correlation coefficients between all three 
    pairs of variables (you can use the pearsonr function from the scipy.stats library). 
2. There is interest in estimating the volume of timber from trees using either the girth or the 
    height of the trees, or both. Perform a regression analysis to decide which of the three 
    possible models you would recommend using. Interpret your results and provide a short 
    conclusion of your findings.

```python
import statsmodels.api as sm
```

```python
df3 = data('trees')
```

```python
df3.head()
```

```python
from scipy.stats import pearsonr
```

```python
# Calculate correlation coefficients and p-values for all pairs of numerical variables
corr_girth_height, p_value_girth_height = pearsonr(df3['Girth'], df3['Height'])
corr_girth_volume, p_value_girth_volume = pearsonr(df3['Girth'], df3['Volume'])
corr_height_volume, p_value_height_volume = pearsonr(df3['Height'], df3['Volume'])

# print correlation coefficients
print(f"Correlation between Girth and Height: {corr_girth_height:.4f}, p-value: {p_value_girth_height:.4f}")
print(f"Correlation between Girth and Volume: {corr_girth_volume:.4f}, p-value: {p_value_girth_volume:.4f}")
print(f"Correlation between Height and Volume: {corr_height_volume:.4f}, p-value: {p_value_height_volume:.4f}")
```

```python
X = sm.add_constant(df3[['Girth', 'Height']])
y = df3['Volume']
```

```python
model = sm.OLS(y, X).fit()

print("\nRegression Analysis Summary:")
print(model.summary())
```

```python
df3.head()
```

```python
sns.regplot(x='Girth', y='Volume', data=df3, scatter_kws={'s': 80})
plt.xlabel('Girth')
plt.ylabel('Volume')
plt.title('Regression Analysis: Volume ~ Girth')
plt.show()
```

```python
sns.regplot(x='Height', y='Volume', data=df3, scatter_kws={'s': 80})
plt.xlabel('Height')
plt.ylabel('Volume')
plt.title('Regression Analysis: Volume ~ Girth')
plt.show()
```
